## ***5-Day LSTM***

```
import matplotlib.pyplot as plt
```

```
NUM_SHOOTINGS = 5
X_TICKS = ['Sandy Hook (2012)', 'San Bernardino (2016)', 'Orlando (2017)', 'Las Vegas (2018)', 'Parkland (2019)']
```

**Initial Surge**

```
y_initial_surge = [4.422018349, 4.378242678, 2.91220062, 3.794277213, 6.045446951]

plt.plot(range(NUM_SHOOTINGS), y_initial_surge, 'o-')
plt.xlabel('Shooting year')
plt.xticks(range(NUM_SHOOTINGS), X_TICKS, rotation=50)
plt.ylabel('Initial surge')
plt.show()
```

***Initial Dropoff***


```
y_initial_dropoff = [0.222018349, 1.656603774, 1.066666667, 2.115572441, 2.733333333]

plt.plot(range(NUM_SHOOTINGS), y_initial_dropoff, 'o-')
plt.xlabel('Shooting year')
plt.xticks(range(NUM_SHOOTINGS), X_TICKS, rotation=50)
plt.ylabel('Initial dropoff')
plt.show()
```

***Initial Dropoff Slope Magnitude***

```
y_initial_dropoff_slope_magnitude = [2.141666667, 0.898485259, 0.866854496, 0.626218134, 0.55]

plt.plot(range(NUM_SHOOTINGS), y_initial_dropoff_slope_magnitude, 'o-')
plt.xlabel('Shooting year')
plt.xticks(range(NUM_SHOOTINGS), X_TICKS, rotation=50)
plt.ylabel('Initial dropoff slope magnitude')
plt.show()
```

***Length of Tail***

```
y_len_tail = [3, 5, 11, 11, 11]

plt.plot(range(NUM_SHOOTINGS), y_len_tail, 'o-')
plt.xlabel('Shooting year')
plt.xticks(range(NUM_SHOOTINGS), X_TICKS, rotation=50)
plt.ylabel('Length of tail')
plt.show()
```
